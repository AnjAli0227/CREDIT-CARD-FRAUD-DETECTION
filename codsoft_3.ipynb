```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
# Load dataset
# Replace 'credit_card_transactions.csv' with the path to your dataset
data = pd.read_csv('/content/drive/MyDrive/Codsoft/fraud credit/creditcard.csv')
```

```python
# Display first few rows of the dataset
print(data.head())
```

```python
# Preprocess data
# Assuming 'Class' is the target variable (fraudulent: 1, genuine: 0)
X = data.drop('Class', axis=1)
y = data['Class']
```

```python
# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Handle class imbalance
# Using SMOTE for oversampling and RandomUnderSampler for undersampling
oversample = SMOTE()
undersample = RandomUnderSampler()
```

```python
# Handle missing values in the target variable
# Options include dropping rows with missing targets or imputing the missing values
# Example: Dropping rows with missing targets
y = y.dropna()

# Adjust X to match the dropped rows in y
X_scaled = X_scaled[y.index]

# Create a pipeline for oversampling and undersampling
pipeline = Pipeline(steps=[('o', oversample), ('u', undersample)])

# Fit and resample the data
X_resampled, y_resampled = pipeline.fit_resample(X_scaled, y)
```

```python
# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
```

```python
# Train model
# You can switch between Logistic Regression and Random Forest by commenting/uncommenting the appropriate lines
# model = LogisticRegression(random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred = model.predict(X_test)
```

```python
# Evaluate model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
